```python
from langchain_pinecone import PineconeVectorStore
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. Initialize the embedding model
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 2. Wrap your Pinecone index
vector_store = PineconeVectorStore(index=index, embedding=embedding)

# 3. Get the retriever
retriever = vector_store.as_retriever()

# 4. Load the LLM
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")  # or another supported LLM

# 5. Create the RetrievalQA chain (acts like a query engine)
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)

# 6. Ask a question
query = "What is the purpose of this document?"
result = qa_chain({"query": query})

# 7. View the result
print(result["result"])
```

```python
import requests



years = list(range(1998, 2025))  # 2025 is exclusive, so this goes up to 2024

base_url = "https://www.berkshirehathaway.com/"
for year in years:
    if year%6 != 0:
      continue;
    report = str(year) + "ar"
    url = base_url + report + "/" + report + ".pdf"
    print(f"Downloading report for year {year} ...")

    response = requests.get(url)
    with open(f"pdfs/{report}.pdf", "wb") as f:
        f.write(response.content)

print("Downloaded reports")
```
